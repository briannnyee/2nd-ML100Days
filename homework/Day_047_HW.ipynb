了解如何使用 Sklearn 中的 hyper-parameter search 找出最佳的超參數

### 作業
請使用不同的資料集，並使用 hyper-parameter search 的方式，看能不能找出最佳的超參數組合

```python
from sklearn import metrics, datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
```

```python
digit = datasets.load_digits()

model = GradientBoostingClassifier(random_state = 50)

x_train, x_test, y_train, y_test = train_test_split(digit.data, digit.target, test_size=0.25, random_state=50)

model_before = model.fit(x_train, y_train)

y_pred = model_before.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ",acc)
```

```python
n_estimators = [100, 200, 300]
max_depth = [1, 3, 5]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

grid_search = GridSearchCV(model, param_grid, scoring="neg_mean_squared_error", n_jobs=-1, verbose=1)

grid_result = grid_search.fit(x_train, y_train)
```

```python
print("Best Accuracy: %f using %s" %(grid_result.best_score_, grid_result.best_params_))
```

```python
clf_bestparam = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],n_estimators=grid_result.best_params_['n_estimators'],random_state=50)

clf_bestparam.fit(x_train,y_train)

y_pred = clf_bestparam.predict(x_test)
```

```python
acc_bestparam = metrics.accuracy_score(y_test, y_pred)
print("BestParam Accuracy: ",acc_bestparam)
```
