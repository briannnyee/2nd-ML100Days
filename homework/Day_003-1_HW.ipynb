### Hints: [隨機產生數值](https://blog.csdn.net/christianashannon/article/details/78867204)

```python
import pandas as pd
import numpy as np
```

```python
data = {'國家': ['US','Taiwan','Japan'],
        '人口': np.random.randint(1000000,size=3)}
data = pd.DataFrame(data)
```

```python
print(data)
```

```python
data.groupby('國家',as_index = False)['人口'].max()
```
